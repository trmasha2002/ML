```python
import cv2
import gensim
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import cv2 
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from skimage import measure
import random
def draw_picture(image, bgr=False):
    b, g, r = cv2.split(image) 
    new_image = cv2.merge([r, g, b])
    plt.figure(figsize=(7, 5))
    plt.axis('off')
    plt.imshow(new_image)
    plt.show()
```

```python
import skimage
from skimage import measure
img = cv2.imread('birds.jpg')
draw_picture(img)
X = []
K = [2, 3, 8, 10, 20]
images = []
size = img.shape
for x in K:
    X = img.reshape(size[0] * size[1], 3).copy()  # преобразование координат, чтобы их можно было закинуть в k-means
    kmeans = KMeans(n_clusters=x, random_state=0).fit(X)
    means = [[0, 0, 0] for i in range(x)] # массив для средних цветов
    counts = [0 for i in range(x)] # количество точек в класстере
    for i in range(len(X)):
        counts[kmeans.labels_[i]] += 1
        for j in range(len(X[i])):
            means[kmeans.labels_[i]][j] = (means[kmeans.labels_[i]][j] + X[i][j]) # считаем сумму всех цветов в одном класстере
    for i in range(len(X)):
        for j in range(len(X[i])):
            X[i][j] = means[kmeans.labels_[i]][j] / counts[kmeans.labels_[i]] # считаем средний цвет класстера
    X = X.reshape(size[0], size[1], 3) # возращаем картинку в трехмерное изображение, чтобы потом не было проблем с SSIM
    images.append(X)
    draw_picture(X) # рисуем картину
for i in range(len(images)):
    print(skimage.measure.compare_ssim(img, images[i], win_size=None, gradient=False, data_range=None, multichannel=True, gaussian_weights=False, full=False))
    # считаем метрику SSIM, сравнивая исходную картинку, с картинкой(с уменьшеным кол-во цветов)
```

```python
from sklearn.cluster import DBSCAN
from skimage import measure
img = cv2.imread('birds.jpg')
draw_picture(img)
X = []
K = [1, 2, 3, 10, 20, 40] # перебираю количество соседей у одного класстера 
images = []
size = img.shape 
for x in K:
    X = img.reshape(size[0] * size[1], 3).copy() # делаю из X двухмерный массив 
    dbscan = DBSCAN(eps=2, min_samples=x).fit(X) # применяю dbscan 
    means = [[0, 0, 0] for i in range(len(X) + 1)] # опять же беру массив для средних значений,  поскольку я не знаю кол-во кластеров беру по длине X
    counts = [0 for i in range(len(X) + 1)] # аналогично с counts 
    for i in range(len(X)):
        counts[dbscan.labels_[i] + 1] += 1 # считаю кол-во в одном класстере
        for j in range(len(X[i])):
            means[dbscan.labels_[i] + 1][j] = (means[dbscan.labels_[i] + 1][j] + X[i][j]) # считаю суммарное кол-во цветов
    for i in range(len(X)):
        for j in range(len(X[i])):
            X[i][j] = means[dbscan.labels_[i] + 1][j] / counts[dbscan.labels_[i] + 1] # узнаю среднее значение 
    X = X.reshape(size[0], size[1], 3) # возращаю матрицу к трехмерному массиву 
    images.append(X) # добавляю в картинки 
    draw_picture(X) # рисую 
for i in range(len(images)):
    print(skimage.measure.compare_ssim(img, images[i], win_size=None, gradient=False, data_range=None, multichannel=True, gaussian_weights=False, full=False))
    # cнова считаю метриек SSIM
```
